```python
# STEP 3: Loading the Dataset
# We use pandas library to read our CSV file

import pandas as pd

# Load the dataset from the file path
df = pd.read_csv("/kaggle/input/datasets/muhammadumer7804/sales-dataset/retail_sales.csv")

# Print a message to confirm it loaded successfully
print("Dataset loaded successfully!")
```

```python
# STEP 5: Finding Missing Values
# This shows exactly how many missing (empty) values are in each column

missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
```

```python
# STEP 6: Fixing Missing Values

# For text columns, fill missing values with the most common value (mode)
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])

# For number columns, fill missing values with the median (middle value)
df['Sales'] = df['Sales'].fillna(df['Sales'].median())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())

print("Missing values have been fixed!")
```

```python
# Let's verify that missing values are gone
print("Missing Values After Fixing:")
print(df.isnull().sum())
```

```python
# STEP 7: Finding Duplicate Rows
# This checks if any row is exactly repeated in the dataset

duplicate_count = df.duplicated().sum()

print("Number of Duplicate Rows Found:", duplicate_count)
```

```python
# STEP 8: Saving the Cleaned Dataset
# This saves our cleaned data as a new CSV file

df.to_csv("cleaned_retail_sales.csv", index=False)

print("Cleaned dataset has been saved successfully!")
```

# Simple Data Cleaning Agent 🤖🧹

## Project Overview
This notebook builds a simple AI agent that automatically cleans a messy retail sales dataset. 
The agent detects missing values, fixes them using appropriate strategies, checks for duplicate 
rows, and saves a clean, ready-to-use version of the dataset.

## Dataset
- **Name:** Retail Sales Dataset
- **Rows:** 1825
- **Columns:** 6 (Date, Category, Sales, Quantity, Profit, Region)

## Steps Performed
1. **Load Dataset** — Imported the CSV file using pandas.
2. **Explore Dataset** — Checked the first few rows, data types, and overall shape.
3. **Detect Missing Values** — Found missing values in 4 columns:
   - Category: 4 missing
   - Sales: 2 missing
   - Quantity: 5 missing
   - Region: 5 missing
4. **Fix Missing Values** —
   - Text columns (`Category`, `Region`) were filled using the **mode** (most frequent value).
   - Numeric columns (`Sales`, `Quantity`) were filled using the **median** (middle value).
5. **Check for Duplicates** — No duplicate rows were found in the dataset.
6. **Save Cleaned Dataset** — Exported the cleaned data as `cleaned_retail_sales.csv`.

## Result
✅ All missing values were successfully fixed
